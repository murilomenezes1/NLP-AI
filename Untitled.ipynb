## Avaliação Intermediária - NLP

### Murilo Menezes

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.feature_extraction.text import CountVectorizer

```

### Leitura do dataset e remoção de dados NaN

```python
df_reviews = pd.read_csv('./olist_order_reviews_dataset.csv').dropna()
df_reviews.head(5)
```

### Recorte de avaliações de 2 estrelas ou menos

```python
bad_reviews = df_reviews[df_reviews['review_score'] <= 2]
bad_reviews
```

## Análise 1: Frequência de tokens unicos nos dados de entrada

### Implementação de stopwords e CountVectorizer

```python

from nltk.corpus import stopwords

stopwords_portuguese = stopwords.words('portuguese')
```

```python
vectorizer = CountVectorizer(max_features=5000, stop_words=stopwords_portuguese, binary=True)
x = vectorizer.fit_transform(bad_reviews['review_comment_message'])
x = np.sum(x, axis=0)
```

### Visualização dos dados

```python
tuplas = [ (x[0,vectorizer.vocabulary_[i]], i) for i in vectorizer.vocabulary_.keys() ]
tuplas_ordenadas = sorted(tuplas, reverse=True) # reverse=True pede uma ordenação em ordem decrescente
palavras = [ t[1] for t in tuplas_ordenadas ]
contagens = [ t[0] for t in tuplas_ordenadas ]

n_palavras = 60
eixo_x = np.arange(n_palavras)
plt.figure(figsize=(14,3))
plt.bar(eixo_x[0:n_palavras], contagens[0:n_palavras])
plt.xticks(eixo_x[0:n_palavras], palavras[0:n_palavras], rotation=70)
plt.show()
```
